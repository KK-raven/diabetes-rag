```python
from langchain_community.document_loaders import PyMuPDFLoader

loader = PyMuPDFLoader("/app/data/raw/package_inserts/metformin.pdf")
pages = loader.load()

# 「社内資料」が含まれるページを探す
for i, page in enumerate(pages):
    if "社内資料" in page.page_content:
        print(f"=== ページ {i} ===")
        print(page.page_content)
        print()
```
