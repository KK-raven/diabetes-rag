```python
from langchain_community.document_loaders import PyMuPDFLoader
from pathlib import Path

def load_pdfs_with_metadata(pdf_dir: Path, doc_type: str) -> list:
    """
    指定フォルダのPDFを全て読み込み、メタデータを付与して返す
    
    doc_type: "package_insert" or "guideline"
    """
    all_documents = []
    pdf_files = sorted(pdf_dir.glob("*.pdf"))
    
    for pdf_path in pdf_files:
        print(f"読み込み中: {pdf_path.name}")
        
        loader = PyMuPDFLoader(str(pdf_path))
        pages = loader.load()
        
        # 各ページにメタデータを追加
        for page in pages:
            page.metadata["doc_type"] = doc_type
            page.metadata["file_name"] = pdf_path.name
            page.metadata["drug_name"] = pdf_path.stem  # 拡張子なしのファイル名
        
        all_documents.extend(pages)
        print(f"  → {len(pages)}ページ読み込み完了")
    
    return all_documents

# 添付文書を読み込む
insert_dir = Path("/app/data/raw/package_inserts")
guideline_dir = Path("/app/data/raw/guidelines")

print("=== 添付文書の読み込み ===")
insert_docs = load_pdfs_with_metadata(insert_dir, "package_insert")

print("\n=== ガイドラインの読み込み ===")
guideline_docs = load_pdfs_with_metadata(guideline_dir, "guideline")

# 合体
all_docs = insert_docs + guideline_docs

print(f"\n=== 完了 ===")
print(f"添付文書: {len(insert_docs)}ページ")
print(f"ガイドライン: {len(guideline_docs)}ページ")
print(f"合計: {len(all_docs)}ページ")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# チャンク分割の設定
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,       # 1チャンクの最大文字数
    chunk_overlap=50,     # 隣り合うチャンクで重複させる文字数
    separators=["\n\n", "\n", "。", "、", ""]  # この順番で区切りを探す
)

# 分割実行
chunks = splitter.split_documents(all_docs)

print(f"分割前: {len(all_docs)}ページ")
print(f"分割後: {len(chunks)}チャンク")
print(f"\n--- チャンクのサンプル ---")
print(f"テキスト:\n{chunks[0].page_content}")
print(f"\nメタデータ:\n{chunks[0].metadata}")
```

```python
# from langchain_openai import OpenAIEmbeddings
# from langchain_chroma import Chroma
# import os

# # Embeddingモデルの設定
# embeddings = OpenAIEmbeddings(
#     model="text-embedding-3-small",
#     api_key=os.getenv("OPENAI_API_KEY")
# )

# # ChromaDBへの登録
# vectorstore = Chroma.from_documents(
#     documents=chunks,
#     embedding=embeddings,
#     persist_directory="/app/data/chroma_db",
#     collection_name="diabetes_rag"
# )

# print(f"登録完了: {vectorstore._collection.count()}チャンク")

from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
import os

embeddings = OpenAIEmbeddings(
    model="text-embedding-3-small",
    api_key=os.getenv("OPENAI_API_KEY")
)

chroma_dir = "/app/data/chroma_db"

# ChromaDBが既に存在する場合は読み込むだけ、なければ新規作成
if os.path.exists(chroma_dir) and os.listdir(chroma_dir):
    print("既存のChromaDBを読み込みます...")
    vectorstore = Chroma(
        persist_directory=chroma_dir,
        embedding_function=embeddings,
        collection_name="diabetes_rag"
    )
    print(f"読み込み完了: {vectorstore._collection.count()}チャンク")
else:
    print("ChromaDBを新規作成します...")
    vectorstore = Chroma.from_documents(
        documents=chunks,
        embedding=embeddings,
        persist_directory=chroma_dir,
        collection_name="diabetes_rag"
    )
    print(f"登録完了: {vectorstore._collection.count()}チャンク")
```

```python
# 検索テスト
query = "メトホルミンの禁忌を教えてください"

results = vectorstore.similarity_search(
    query=query,
    k=3  # 上位3件を取得
)

print(f"検索クエリ: {query}\n")
for i, doc in enumerate(results):
    print(f"--- 結果 {i+1} ---")
    print(f"テキスト:\n{doc.page_content}")
    print(f"\nメタデータ: {doc.metadata}")
    print()
```

```python
# メトホルミンのチャンクを全件取得して禁忌を含むものを探す
all_metformin = vectorstore.get(
    where={"drug_name": "metformin"},
)

print(f"メトホルミンのチャンク数: {len(all_metformin['documents'])}")
print()

# 禁忌を含むチャンクを探す
kinki_chunks = [
    doc for doc in all_metformin["documents"]
    if "禁忌" in doc
]

print(f"「禁忌」を含むチャンク数: {len(kinki_chunks)}")
print()

if kinki_chunks:
    print("--- 禁忌を含む最初のチャンク ---")
    print(kinki_chunks[0])
```

```python
from langchain_openai import ChatOpenAI
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
import os

# LLMの設定
llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL", "gpt-4o-mini"),
    temperature=0,
    api_key=os.getenv("OPENAI_API_KEY")
)

# プロンプトテンプレートの設定
prompt_template = """
あなたは医薬品の専門家です。
以下の医薬品添付文書およびガイドラインの情報をもとに質問に回答してください。

提供された情報に含まれていない内容については「提供された文書には記載がありません」と答えてください。
回答の最後に必ず参照した文書名を記載してください。

【参照情報】
{context}

【質問】
{question}

【回答】
"""

prompt = PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)

# RAGチェーンの構築
retriever = vectorstore.as_retriever(
    search_kwargs={"k": 5}
)

rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    chain_type_kwargs={"prompt": prompt},
    return_source_documents=True
)

print("RAGチェーン構築完了")
```

```python
def ask(query: str):
    """RAGチェーンに質問して結果を表示する"""
    result = rag_chain.invoke({"query": query})
    
    print(f"【質問】{query}")
    print(f"\n【回答】")
    print(result["result"])
    print(f"\n【参照チャンク】")
    for i, doc in enumerate(result["source_documents"]):
        print(f"\n  --- チャンク{i+1} ---")
        print(f"  ファイル：{doc.metadata.get('file_name')}")
        print(f"  種別：{doc.metadata.get('doc_type')}")
        print(f"  ページ：{doc.metadata.get('page')}")
        print(f"  内容（先頭100文字）：{doc.page_content[:100]}")

# まず1つ試す
ask("メトホルミンの禁忌を教えてください")
```

```python
# 様々な種類の質問を試す
questions = [
    "メトホルミンの用法・用量を教えてください",
    "グリメピリドの副作用を教えてください",
    "SGLT2阻害薬の特徴を教えてください",
    "腎機能が低下している患者に使用できる糖尿病薬は何ですか",
    "インスリンとGLP-1受容体作動薬の違いを教えてください"
]

for q in questions:
    print("="*60)
    ask(q)
    print()
```
